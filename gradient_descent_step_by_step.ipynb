```python
from sklearn.datasets import make_regression
```

```python
import numpy as np
```

```python
X,y=make_regression(n_samples=4,n_features=1,n_informative=1,n_targets=1,noise=80,random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
```

```python
# Applying OLS(Ordinary Least Square)
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
```

```python
reg.fit(X,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red')
```

```python
# Applying Gradient Descent Method assuming slope is constant m=78.35
# and assuming the starting value for intercept b=0
y_pred=((78.35*X)+0).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred,color='green',label='b=0')
plt.legend()
```

```python
m=78.35
b=0

loss_slope=-2*np.sum(y-m*X.ravel()-b)
loss_slope
```

```python
# taking learning rate = 0.1
lr=0.1
step_size=loss_slope*lr
step_size
```

```python
# calculate new b
b=b-step_size
```

```python
b
```

```python
y_pred1=((78.95*X)+b).reshape(4)

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred1,color='green',label='b={}'.format(b))
plt.plot(X,y_pred,color='blue',label='b = 0')
plt.legend()
plt.show()
```

```python
# Iteration 2

loss_slope=-2*np.sum(y-m*X.ravel()-b)
loss_slope
```

```python
step_size=loss_slope*lr
step_size
```

```python
b=b-step_size
```

```python
b
```

```python
y_pred2=((78.95*X)+b).reshape(4)
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred2,color='green',label='b={}'.format(b))
plt.plot(X,y_pred1,color='blue',label='b={}'.format(b))
plt.plot(X,y_pred,color='yellow',label='b=0')
plt.legend()
plt.show()
```

```python
# Iteration 3
loss_slope=-2*np.sum(y-m*X.ravel()-b)
loss_slope
```

```python
step_size=loss_slope*lr
step_size
```

```python
b=b-step_size
```

```python
b
```

```python
y_pred3=((78.95*X)+b).reshape(4)
```

```python
y_pred3
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='Black',label='OLS')
plt.plot(X,y_pred3,color='red',label='b={}'.format(b))
plt.plot(X,y_pred2,color='blue',label='b={}'.format(b))
plt.plot(X,y_pred1,color='green',label='b={}'.format(b))
plt.plot(X,y_pred,color='yellow',label='b=0')
plt.legend()
```

```python
# doing this using a loop

b=-100
m=78.35
lr=0.1

epochs=10
plt.plot(X,reg.predict(X),label='OLS')
for i in range(epochs):
  loss_slope=-2*np.sum(y-m*X.ravel()-b)
  b=b-(lr*loss_slope)

  y_pred=m*X+b
  plt.plot(X,y_pred)
plt.scatter(X,y)
plt.legend()
```
